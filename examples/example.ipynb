# 3 Building co-occurence matrix using context 

## 3.A: Using all vocabularies as context

#### Only lemma 

```python
stopwords_filter = mangoes.corpus.remove_elements(nltk.corpus.stopwords.words('english'))

# Only keep lemma (since we only keep lemma thus stopwords_filter does not need attribute)
lemma_vocabulary = corpus.create_vocabulary(attributes=("lemma"), filters = [ stopwords_filter, mangoes.corpus.remove_most_frequent(100),mangoes.corpus.remove_least_frequent(2)])
print("Lemma vocabulary :", len(lemma_vocabulary))

coocc_count = mangoes.counting.count_cooccurrence(corpus, lemma_vocabulary, context=lemma_vocabulary)
coocc_count.pprint(display=display)
```

```python
i = 30
lemma_vocabulary[i], coocc_count.get_closest_words(lemma_vocabulary[i], 10)
```

```python
lemma_vocabulary.word_index["anyone"]
```

#### Lemma and POS

```python
stopwords_filter = mangoes.corpus.remove_elements(nltk.corpus.stopwords.words('english'),attribute=("lemma"))


lemma_POS_vocabulary = corpus.create_vocabulary(attributes=( "lemma","POS"), filters = [ stopwords_filter, mangoes.corpus.remove_most_frequent(100),mangoes.corpus.remove_least_frequent(2)])

#print("Lemma vocabulary :", len(lemma_vocabulary))

coocc_count = mangoes.counting.count_cooccurrence(corpus, lemma_POS_vocabulary, context=lemma_POS_vocabulary)
coocc_count.pprint(display=display)
```

```python
stopwords_filter_lemma = mangoes.corpus.remove_elements(nltk.corpus.stopwords.words('english'),attribute=("lemma"))
stop_words_filter_form = mangoes.corpus.remove_elements(nltk.corpus.stopwords.words('english'),attribute=("form"))

token_vocabulary = corpus.create_vocabulary(attributes=("form", "lemma","POS"), filters = [ stopwords_filter_lemma, stop_words_filter_form , mangoes.corpus.remove_most_frequent(100),mangoes.corpus.remove_least_frequent(2)])

#print("Lemma vocabulary :", len(lemma_vocabulary))

coocc_count = mangoes.counting.count_cooccurrence(corpus, token_vocabulary, context=token_vocabulary)
coocc_count.pprint(display=display)
```

## 3.B context using Windows
#### Building Vocabulary 

```python
stopwords_filter = mangoes.corpus.remove_elements(nltk.corpus.stopwords.words('english'))

# Only keep lemma (since we only keep lemma thus stopwords_filter does not need attribute)
lemma_vocabulary = corpus.create_vocabulary(attributes=("lemma"), filters = [ stopwords_filter, mangoes.corpus.remove_most_frequent(100),mangoes.corpus.remove_least_frequent(2)])
```

```python
simple_window = mangoes.context.Window(size=7, vocabulary=lemma_vocabulary)
simple_window
print(f"context = symmetric window of size 5")
coocc_count = mangoes.counting.count_cooccurrence(corpus,  
                                                    lemma_vocabulary, 
                                                    context=simple_window)
coocc_count.pprint(display=display)
coocc_count.shape
```

## 3.C Context using Dependency relation

```python
# always set punctuation to False since dependency relation uses ID 
dependency_corpus = mangoes.Corpus(UD_English_EWT, reader=mangoes.corpus.CONLLU, language="English", lower=True, ignore_punctuation=False)

# vocabulary attributes/entity need to be matched 
stopwords_filter = mangoes.corpus.remove_elements(nltk.corpus.stopwords.words('english'))
freq_vocabulary = dependency_corpus.create_vocabulary(attributes=("lemma","POS","form"), filters = [ stopwords_filter, mangoes.corpus.remove_most_frequent(100),mangoes.corpus.remove_least_frequent(2)])

dependency_context = mangoes.context.DependencyBasedContext(entity=("lemma","POS","form"), labels=True,collapse=True, vocabulary=freq_vocabulary)

coocc_count = mangoes.counting.count_cooccurrence(dependency_corpus,  
                                                freq_vocabulary, 
                                                context=dependency_context,
                                                )

coocc_count.pprint(display=display)
```

```python
coocc_count.shape
```

```python
stopwords_filter = mangoes.corpus.remove_elements(nltk.corpus.stopwords.words('english'))
freq_vocabulary = dependency_corpus.create_vocabulary(attributes=("form"), filters = [ stopwords_filter, mangoes.corpus.remove_most_frequent(100),mangoes.corpus.remove_least_frequent(2)])

dependency_context = mangoes.context.DependencyBasedContext(entity=("form"), labels=True,collapse=True, vocabulary=freq_vocabulary)

coocc_count = mangoes.counting.count_cooccurrence(dependency_corpus,  
                                                freq_vocabulary, 
                                                context=dependency_context,
                                                )

coocc_count.pprint(display=display)
```

```python
stopwords_filter = mangoes.corpus.remove_elements(nltk.corpus.stopwords.words('english'))
freq_vocabulary = dependency_corpus.create_vocabulary(attributes=("form", "POS"), filters = [ stopwords_filter, mangoes.corpus.remove_most_frequent(100),mangoes.corpus.remove_least_frequent(2)])

dependency_context = mangoes.context.DependencyBasedContext(entity=("form","POS"), labels=True,collapse=True, vocabulary=freq_vocabulary)

coocc_count = mangoes.counting.count_cooccurrence(dependency_corpus,  
                                                freq_vocabulary, 
                                                context=dependency_context,
                                                )

coocc_count.pprint(display=display)
```

## Applying weight function

```python
import mangoes.weighting
ppmi_representation = mangoes.create_representation(coocc_count, weighting=mangoes.weighting.PPMI())
```

```python
ppmi_representation.pprint(display=display)
```

```python
ppmi_representation.get_closest_words(freq_vocabulary[1], 10)
```

```python
import mangoes.reduction
embeddings = mangoes.create_representation(ppmi_representation, reduction=mangoes.reduction.PCA(dimensions=300))
```
